```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
```

Basic Graph

```python
x=[1,2,3]
y = [2,4,6]

plt.figure(figsize=(6,4) ,dpi=100) # resize your graph

plt.plot( x, y , label='2X' , color='red' , linewidth = 2 , marker ='.' ,linestyle = '--' , markersize = 10 ,markeredgecolor = 'blue' , markeredgewidth = 3 , markerfacecolor = 'yellow' , alpha = 0.5)  

plt.title('Our First Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

Bar Chart

```python
plt.figure(figsize=(9,5) )
lable =['a' , 'b' ,  'c']
values =[1,4,2] 

bar = plt.bar(lable , values ,width=0.4 ,color=['red' , 'green' , 'blue'] , edgecolor='black' ) 
plt.show()
```

Scatter plot

```python
plt.figure(figsize=(9,5) )
x_data = np.random.random(50)*100
y_data = np.random.random(50)*100

plt.scatter(x_data ,y_data ,color = 'green' , s=100  ,alpha=0.8 , edgecolor = 'red' , linewidth = 2 , marker = 'o' )

plt.show()
```

Line Plots

```python
plt.figure(figsize=(9,5) )
years = [2006 * x for x in range(16)]
weights =[80 ,83 , 84 , 85, 86 , 82 , 81 , 79 , 83 , 80 , 82 , 82 , 83 , 81 ,80 ,79]

plt.plot(years ,weights ,marker='.' , markersize=10 , c='r' ,mec='blue' ,lw=2 ,linestyle='--' , alpha=0.5 )
plt.show()
```

Histogram

```python
plt.figure(figsize=(9,5) )
ages = np.random.normal(20 ,1.5 ,1000)
plt.hist(ages ,bins=30 ,edgecolor='black' , linewidth=1 , color='green' , alpha=0.5 ,  rwidth=0.8 ,  density=True ,  orientation='vertical')
plt.show()
```

Pie chart ( catigorical data)

```python
plt.figure(figsize=(9,5) )
langs =['python' , 'c++' , 'Java' ,'go' , 'C#']
votes = [50 , 24 , 14 , 6 , 17 ]
style.use('ggplot')

plt.pie(votes , labels=langs , autopct='%1.1f%%' , shadow=True , explode=(0.1,0,0,0,0) , startangle=90 , radius=1.2 , counterclock=False , wedgeprops={'linewidth':2} , textprops={'fontsize':15} , normalize=True) 


plt.show()
```

BoxPlots

```python
plt.figure(figsize=(9,5) )
heights = np.random.normal(172,8,300)

plt.boxplot(heights , patch_artist=True , boxprops={'color':'red'} , medianprops={'color':'black'} , whiskerprops={'color':'green'} , capprops={'color':'blue'} , showmeans=True)


plt.show()
```

Multiple Plots

```python
plt.figure(figsize=(10,7) )
Stock_a =[100 , 102 , 99 , 101  , 101 , 100 ,102 ]
Stock_b =[90 , 95 , 102 , 104 , 105 , 103 , 109 ]
Stock_c =[110 , 115 , 100 , 105 , 100 ,98 , 95 ]

plt.plot(Stock_a , label='company1' , color='red' , linewidth = 2 , marker ='.' ,linestyle = '--' , markersize = 10 ,markeredgecolor = 'blue' , markeredgewidth = 3 , markerfacecolor = 'yellow' , alpha = 0.5)
plt.plot(Stock_b , label='company2' , color='green' , linewidth = 2 , marker ='.' ,linestyle = '--' , markersize = 10 )
plt.plot(Stock_c , label='company3' , color='blue', linewidth = 2 , marker ='.' ,linestyle = '--')


plt.legend(loc='lower center')
plt.show()
```
